**GrÃ¡fico boxplot**

```python
# Exemplo de codigo para grafico boxplot.
import matplotlib.pyplot as plt
import numpy as np
import statistics as sta

lista = [10,20,20,24,28,28,30,32,32,32,36,40,43,56,60]

mediana = np.median(lista)
media = np.average(lista)
moda = sta.mode(lista)
q1 = np.percentile(lista, 25, method="midpoint")
q3 = np.percentile(lista, 75, method="midpoint")
I = (q3 - q1)
min = q1 - 1.5*I
max = q3 + 1.5*I

outlier = []
for i in range(len(lista)):
  if lista[i] > max or lista[i] < min:
    outlier.append(lista[i])

print(f"mediana = {mediana}")
print(f"media = {media}")
print(f"moda = {moda}")
print(f"1Â° quartil = {q1}")
print(f"3Â° quartil = {q3}")
print(f"I = {I}")
print(f"min = {min}")
print(f"max = {max}")
print(f"Outlier ={outlier}")

plt.boxplot(lista, vert = False)
#plt.savefig("grafico boxplot.png")
plt.show()
```

**GrÃ¡fico de Barras**

```python
# exemplo de codigo para grafico de barras
import matplotlib.pyplot as plt
import numpy as np

x = ["Vira-lata", "Golden", "Shitzu", "Labrador", "Pitbull"]
y = [25, 13, 7, 17, 21]

cores = ["red", "yellow", "silver", "black", "blue"]

plt.bar(x, y, color = cores)
plt.title("RaÃ§as de Cachorro adotados em janeiro", color= "black")
plt.ylabel("Quantidade", color= "black")
plt.xlabel("RaÃ§as de cachorro", color="black")
plt.ylim([0,30])

for i, val in enumerate (y):
  texto = f"{val} "
  y_coord = val + 1
  x_coord = i - 0.1
  plt.text(x=x_coord, y=y_coord, s=texto, fontsize=10)

plt.show()
#plt.savefig("Grafico de barras.png)
```

**GrÃ¡fico de linhas**

```python
# exemplo de codigo de grafico de linhas
import matplotlib.pyplot as plt
import numpy as np

acaoA = [30,70,190,110,230,310,350,150,350,390,410,400]
acaoB = [100,250,400,200,350,500,150,300,350,370,320,330]
mes = ["janeiro","fevereiro","marÃ§o","abril","maio","junho","julho","agosto","setembro","outubro","novembro","dezembro"]


plt.figure(figsize=(15, 6))

for i, val in enumerate(acaoA):
  texto = f"{val} "
  x_coord = i - 0.3
  y_coord = val + 4
  plt.text(x=x_coord, y=y_coord, s=texto, fontsize= 10)

for i, val in enumerate(acaoB):
  texto = f"{val} "
  x_coord = i -0.3
  y_coord = val + 4
  plt.text(x=x_coord, y=y_coord, s=texto, fontsize= 10)


plt.plot(mes, acaoA,
         marker = "o",
         color = "red",
         linestyle = "--",
         linewidth = 2,
         label = "AÃ§Ã£o A",)
plt.plot(mes, acaoB,
         marker = "o",
         color = "blue",
         label = "AÃ§Ã£o B")


plt.title("AÃ§Ã´es no ano de 2023", color ="black")
plt.ylim([0,700])
plt.legend(loc = "upper left")
plt.grid()
#plt.figure(facecolor ="brown")
#plt.axes().set_facecolor("#FFE7BA")
#plt.savefig("grafico de linhas.png")
plt.show()
```
